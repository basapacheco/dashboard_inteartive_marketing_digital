# Análise de Dados de Marketing Digital

Este notebook realiza a análise exploratória de dados (EDA) para entender melhor a estrutura e as tendências dos dados de Google Analytics e Facebook Ads.

## Importar e Explorar os Dados

```python




````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Carregar os dados do Google Analytics e Facebook Ads
ga_data = pd.read_json('../data/google_analytics_data.json')
fb_data = pd.read_json('../data/facebook_ads_data.json')
````

# Análise Exploratória de Dados (EDA)

````python
print("Google Analytics Data:")
print(ga_data.describe())
print(ga_data.head())
````

````python
print("\nFacebook Ads Data:")
print(fb_data.describe())
print(fb_data.head())
````

````python
# Visualizar algumas métricas
ga_data['sessions'].plot(kind='line', title='Sessões do Google Analytics')
plt.show()

fb_data['impressions'].plot(kind='line', title='Impressões dos Anúncios do Facebook')
plt.show()
````
