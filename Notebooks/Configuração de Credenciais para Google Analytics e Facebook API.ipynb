```python
import json
import facebook
import requests
import random
from datetime import datetime, timedelta

# Carregar as credenciais do arquivo config.json
with open('config.json') as config_file:
    config = json.load(config_file)

access_token = config['facebook']['access_token']

# Configurar o cliente do Facebook
graph = facebook.GraphAPI(access_token=access_token, version="3.1")

def get_facebook_ads_data():
    # Substituir pelo ID da sua conta de anúncios
    ad_account_id = 'act_2128826710825812'
    
    try:
        # Fazer uma solicitação à API do Facebook
        ads = graph.get_connections(id=ad_account_id, connection_name='ads')
        
        if not ads['data']:
            raise facebook.GraphAPIError("Nenhum anúncio encontrado. Gerando dados fictícios.")
        
        ads_data = []
        for ad in ads['data']:
            ad_id = ad['id']
            ad_insights = graph.get_connections(id=ad_id, connection_name='insights', fields='impressions,clicks,ctr')
            ad_data = {
                'ad_id': ad_id,
                'campaign_name': ad.get('campaign_name', 'N/A'),
                'impressions': ad_insights['data'][0].get('impressions', '0'),
                'clicks': ad_insights['data'][0].get('clicks', '0'),
                'ctr': ad_insights['data'][0].get('ctr', '0')
            }
            ads_data.append(ad_data)
        
        return ads_data
    except facebook.GraphAPIError as e:
        print(f"Erro ao obter os dados dos anúncios: {e}")
        return generate_dummy_data()

def generate_dummy_data():
    campaigns = ['Campanha Fictícia 1', 'Campanha Fictícia 2', 'Campanha Fictícia 3']
    dummy_data = []

    start_date = datetime.now() - timedelta(days=30)
    for i in range(100):  # Generar 100 registros fictícios
        campaign_name = random.choice(campaigns)
        date = (start_date + timedelta(days=random.randint(0, 30))).strftime('%Y-%m-%d')
        impressions = random.randint(500, 5000)
        clicks = random.randint(50, 500)
        ctr = f'{clicks / impressions * 100:.2f}%'
        
        ad_data = {
            'ad_id': str(random.randint(100000, 999999)),
            'campaign_name': campaign_name,
            'date': date,
            'impressions': impressions,
            'clicks': clicks,
            'ctr': ctr
        }
        dummy_data.append(ad_data)
    
    return dummy_data

# Exemplo de uso da função
if __name__ == "__main__":
    ads_data = get_facebook_ads_data()
    if ads_data:
        with open('data/facebook_ads_data.json', 'w') as f:
            json.dump(ads_data, f, indent=4)
        print(json.dumps(ads_data, indent=4))
    else:
        print("Nenhum dado de anúncio obtido.")
```
